Import required libraries

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

Load training data to a pandas dataframe

```python
df = pd.read_csv('training_data.csv')
```

```python
df.head()
```

```python
#Drop disease column
df.drop('prognosis', axis=1, inplace=True)
```

```python
freq_symptoms = fpgrowth(df, min_support=0.021, max_len=17, use_colnames=True, verbose=1)
freq_symptoms.head()
```

```python
rules = association_rules(freq_symptoms, metric="confidence", min_threshold=0.3)
rules
```

```python
rules.to_csv('rules.csv', index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```

```python
fit = np.polyfit(rules['lift'], rules['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules['lift'], rules['confidence'], 'yo', rules['lift'], fit_fn(rules['lift']))
```

```python
plt.scatter(rules['support'], rules['lift'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('lift')
plt.title('Support vs Lift')
plt.show()
```

Get features similar to headache

```python
related = rules[rules['antecedents']=={'headache'}]
st = list(map(set, related['consequents'].to_list()))
new = set()
for i in st:
  for j in i:
    new.add(j)
new
```

Store related symtoms for each symptom in a dictionary

```python
related_symptoms = {}
symptoms = df.columns
for symptom in symptoms:
  related = rules[rules['antecedents']=={symptom}]
  st = list(map(set, related['consequents'].to_list()))
  new = set()
  for i in st:
    for j in i:
      new.add(j)
  related_symptoms[symptom]=new
```

```python
related_symptoms
```

```python
related_df = pd.DataFrame(related_symptoms.items(), columns=['symptom','related'])
related_df
```

```python
related_df.to_csv('related.csv',index=False)
```
