```
import pandas as pd
import numpy as np
```

Load data for training the model

```
df = pd.read_csv('training_data.csv')

df.head()
```

```
X = df.iloc[:, :-1]
y = df['prognosis']
```

Split the data for training and testing

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

Train the model using random forest classifier

```
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=200,min_samples_leaf=3,max_features=100, max_depth=10)
```

```
rf_clf.fit(X_train, y_train)

accuracy = rf_clf.score(X_test,y_test)
print("Accuracy:", accuracy)
```

Calculate performance measures

```
from sklearn.metrics import classification_report

y_true = y_test
y_pred = rf_clf.predict(X_test)
disease_names = df['prognosis'].unique()
```

```
print(classification_report(y_true, y_pred, target_names=disease_names))
```

Create a dictionary with symptoms as keys and indices as values

```
symptoms_dict = {}

for index, symptom in enumerate(X):
    symptoms_dict[symptom] = index
```

```
symptoms_dict
```

Predict for sample data

```
input_vector = np.zeros(len(symptoms_dict))
```

```
input_vector[[symptoms_dict['cramps'], symptoms_dict['cough']]] = 1
```

```
rf_clf.predict_proba([input_vector])
```
